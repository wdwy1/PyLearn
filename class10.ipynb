```python
from pylab import *

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
show()
```

```python
x=np.arange(-10,10,0.1);
y1=np.sin(x)
y2=np.cos(x)
line=plt.plot(x,y1,x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title("sin&&cos")
plt.xlim(-2,2)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2,
                 np.pi],  [r'$-\pi$', r'$-\pi/2$',
                 r'$0$', r'$\pi/2$', r'$\pi$'])

plt.ylim(-1.25,1.25)
plt.legend((line),('sin','cos'))
plt.grid()
plt.axhline(0)
plt.axvline(0)
plt.axhspan(0,2, alpha=0.3)
plt.axvspan(0,2, alpha=0.3)
plt.annotate('min',[-np.pi/2,-1],[-np.pi/2,-0.5],arrowprops=dict(arrowstyle='fancy'))
plt.text(-2,0,'this is a line')
```
